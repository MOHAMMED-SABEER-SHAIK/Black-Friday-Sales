```python
#Analysing number of products bought by each age group.
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
    
data = pd.DataFrame(lst , columns = ['Age','Products'])
```

```python
black_friday_data = pd.read_csv('BlackFriday.csv')
df = pd.read_csv('BlackFriday.csv')
# Selecting relevant numeric columns for the pair plot
selected_columns = ['Purchase', 'Product_Category_1', 'Product_Category_2']

# Dropping rows with missing values for the selected columns
black_friday_clean = black_friday_data[selected_columns].dropna()

sns.pairplot(black_friday_clean)
plt.show()
```

```python
# Selecting relevant numeric columns for the heatmap
selected_columns = ['Purchase', 'Occupation', 'Product_Category_1', 'Product_Category_2', 'Product_Category_3']

# Dropping rows with missing values for the selected columns
black_friday_clean1 = black_friday_data[selected_columns].dropna()

# Generating the correlation matrix
correlation_matrix = black_friday_clean1.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Selected Features')
plt.show()
```
